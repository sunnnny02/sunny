```python
import re

def extract_administrative_district(address):
    match = re.search(r'(\S*동)', address)  # 주소에서 '동'이 들어있는 부분 추출
    if match:
        return match.group(1)  # 추출된 부분 반환
    else:
        return None

# Apply the function to extract 행정동 and create a new column
df['행정동'] = df['지번 주소'].apply(lambda x: extract_administrative_district(x) if isinstance(x, str) else None)
df['행정동'] = df['행정동'].map(similar_dong_mapping).fillna(df['행정동'])

df.head(30)
```
