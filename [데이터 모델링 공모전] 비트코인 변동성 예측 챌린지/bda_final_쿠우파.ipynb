```python
import pandas as pd
import numpy as np
import os
```

```python
data=pd.read_csv('finaldata.csv') #23.08.01~24.01.27 의 dataframe
sample=data[['time', 'Open', 'High', 'Low', 'Close']] 
sample['time'] = pd.to_datetime(sample['time'])
sample
```

```python
sample['time'] = pd.to_datetime(sample['time'])
sample['hour'] = sample['time'].dt.hour
sample['date'] = sample['time'].dt.date
```

```python
sample1 = sample[['Close']]
sample1.Close = sample1.Close.ffill()
close_data = sample1['Close'].to_numpy()

# Reshape the data into a 2D array with 60 rows and 72(24*3) columns
reshaped_data = close_data.reshape((60, 72))
reshaped_df = pd.DataFrame(reshaped_data)
reshaped_df.columns = range(72)

# Display the reshaped DataFrame
reshaped_df.head()
```

```python
x1 =reshaped_df.iloc[0:-1].to_numpy() 
x1
```

```python
y1 =reshaped_df.iloc[1:].to_numpy()
y1_values=[]
for i in range(72):
    y1_i = y1[:,i]
    y1_values.append(y1_i)
x1_public =reshaped_df.iloc[-1].to_numpy()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=100,random_state=2020,max_features=12) #grid search result
```

```python
predictions1 = []
for y1_value in y1_values :
    rf_regressor.fit(x1,y1_value)
    prediction = rf_regressor.predict(np.expand_dims(x1_public,0))
    predictions1.append(prediction[0])
predictions1 #1.28~1.30 예측한 것 
```
